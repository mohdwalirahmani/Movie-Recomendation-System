```python
import numpy as np
import pandas as pd
import difflib                                                  # this checks the spelling mistakes that user makes while writing input
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Loading data and Pre-Processing


```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# selectiong the relevant features for recomendation
selecting_features = ['genres','keywords','tagline','cast','director']
print(selecting_features)
```

```python
# replacing the null values with null string
for feature in selecting_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the five selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']

# Hum alag-alag text fields ko mila ke ek string bana rahe hain
# Taaki us combined text ka vector ban sake
# Fir vector ke basis pe cosine similarity nikal sakein
# Aur finally similar movies recommend kar saken
```

```python
print(combined_features)
```

```python
# converting text data into feature vectors
vectorizer = TfidfVectorizer()
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

we r doing this bcoz we want to implement cosine similarity algo

```python
# Now we will get the similarity score using cosine similarity
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user
movie_name = input('Enter your favourite movie name: ')
```

```python
# now we will create a list of all the movie name given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# Now we find the close match for the mvoie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of movie with title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting the list of similarity scores
similarity_score = list(enumerate(similarity[index_of_the_movie]))      # enumerate list ko iterate karta hai, aur har item ke saath uska index bhi return karta hai.
similarity_score = [(i,(float(score))) for i, score in similarity_score]
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting movie based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# now we will print the name of similar movies based on index
print("Movies recommended for u are: \n")

i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<=20):
    print(i, '-', title_from_index)
    i+=1
```

##Movie Recomendation System

```python
movie_name = input('Enter your favourite movie name: ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

if not find_close_match:
    print("Sorry, no matching movie found.")
    exit()

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score = [(i,(float(score))) for i, score in similarity_score]

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print("Movies recommended for u are: \n")

i=1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<=20):
    print(i, '-', title_from_index)
    i+=1
```
